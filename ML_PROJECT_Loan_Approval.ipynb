```python
import numpy as np

# Calculate the first quartile (Q1) and third quartile (Q3) of the data
Q1=df.total_income.quantile(.25)# Q1 will contain the value below which 25% of the values in the total_income column fall.
Q3=df.total_income.quantile(.75)# Q3 will contain the value below which 75% of the values in the total_income column fall. 
Q1,Q3

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1


# Calculate the lower and upper bounds for outliers
lower_limit = Q1 - (1.5 * IQR)
upper_limit = Q3 + (1.5 * IQR)
lower_limit
upper_limit
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,random_state=1)
```

```python
#PCA Principle Component Analysis
# from sklearn.decomposition import PCA
# pca=PCA(n_components=3)
# X_train_pca=pca.fit_transform(X_train)
# X_test_pca=pca.transform(X_test)
```

```python
Accuracy_values=[]
Models=[]
```

```python
#K-NEIGHBOURS
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
Models.append('K-Neighbours')
y_pred
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)*100
Accuracy_values.append(acc)
```

```python
#NAIVE BAYES
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred1=nb.predict(X_test)
y_pred1
Models.append('NAIVE BAYES')

acc=accuracy_score(y_test,y_pred1)*100
Accuracy_values.append(acc)
```

```python
#MultinomialNB
from sklearn.naive_bayes import MultinomialNB
Mnb=MultinomialNB()
Mnb.fit(X_train,y_train)
y_pred9=Mnb.predict(X_test)
y_pred9
Models.append('MultinomialNB')
acc=accuracy_score(y_test,y_pred9)*100
Accuracy_values.append(acc)
```

```python
#sUPPORT VECTOR MECHANISM
from sklearn.svm import SVC
svm=SVC()
svm.fit(X_train,y_train)
y_pred2=svm.predict(X_test)
y_pred2
Models.append('SUPPORT VECTOR MECHANISM')
acc=accuracy_score(y_test,y_pred2)*100
Accuracy_values.append(acc)
```

```python
#DECISION TREE
from sklearn.tree import DecisionTreeClassifier
Dclf=DecisionTreeClassifier()
Dclf.fit(X_train,y_train)
y_pred3=Dclf.predict(X_test)
y_pred3
Models.append('DECISION TREE')
acc=accuracy_score(y_test,y_pred3)*100
Accuracy_values.append(acc)
```

```python
#RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
Rclf=RandomForestClassifier(n_estimators=550)
Rclf.fit(X_train,y_train)
y_pred4=Rclf.predict(X_test)
y_pred4
Models.append('RANDOM FOREST')
acc=accuracy_score(y_test,y_pred4)*100
Accuracy_values.append(acc)
```

```python
#LOGISTIC REGRESSSION
from sklearn.linear_model import LogisticRegression
lnr=LogisticRegression()
lnr.fit(X_train,y_train)
y_pred5=lnr.predict(X_test)
y_pred5
Models.append('LOGISTIC REGRESSSION')
acc=accuracy_score(y_test,y_pred5)*100
Accuracy_values.append(acc)
```

```python
from sklearn.model_selection import GridSearchCV
knn1=KNeighborsClassifier()
params={'n_neighbors':[3,5,7,9],'weights':['uniform','distance']}
clf=GridSearchCV(knn1,params,cv=10,scoring='accuracy')
clf.fit(X_train,y_train)
```

```python
print(clf.best_params_)
```

```python
knn2=KNeighborsClassifier(n_neighbors=9,weights='uniform')
knn2.fit(X_train,y_train)
y_pred6=knn2.predict(X_test)
acc=accuracy_score(y_test,y_pred6)*100
Models.append('knn2')
Accuracy_values.append(acc)
```

```python

# create a bar plot of accuracy values for different models
plt.figure(figsize=[8, 4], dpi=100)
plt.title('Accuracy comparison')
plt.xlabel('Accuracy')
plt.ylabel('Algorithm')
sns.barplot(x=Accuracy_values, y=Models)
plt.show()
```
